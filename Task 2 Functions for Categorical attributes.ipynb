```python
list_of_countries= ['AT','BE','BG','CH','DE','DK','ES','FI','FR','GB','GR','HU','IE','IS','IT','LU','NO','NL','PO','PT','RO','RU','SE','TR','UA']
```

```python
# Making an unbalanced categorical atribute of countries
len(list_of_countries)
# 1 country with a probability of 35%, another with 25%, 20%, 19% 
# and the rest of the 21 countries will have a 1% between themselfs
country_35=list_of_countries[0]
country_25=list_of_countries[2]
country_20=list_of_countries[1]
country_19=list_of_countries[3]
others_1 = list_of_countries[4:25]
```

```python
#Making a categorical attribute by building a list of countries into a pandas dataframe
def p():
    return randint(0,100)/100

countries =[]
for i in range(100):
    per = p()
#     print(per)
    if per>0.65:countries.append(country_35)
    elif 0.65>=per>0.40:countries.append(country_25)
    elif 0.40>=per>0.20:countries.append(country_20)
    elif 0.20>=per>0.01:countries.append(country_19)
    else: countries.append(others_1[randint(0,20)])

assert len(countries) == 100
country_attributes = list(set(countries))
countries_dataframe = pd.DataFrame()
countries_dataframe['Country'] = se = pd.Series(countries)
list(country_attributes)
```

a) Write a function with 2 input arguments i) the values of a categorical attribute like the values of Country in the above dataset ii) a percentage between 0%-100%. The output should be as the first input with the values that occur less frequent than the input percentage replaced by a generic categorical value. 

```python
# This method corresponds to solution of a) by replacing less frequent values by 
# a generic attribute
def naive_dimensionality_reduction(countries_dataframe, percentage):
    frequency_dictionary=dict((value,0) for value in country_attributes)
    for i in countries_dataframe.values:
        for k in frequency_dictionary.keys():
            if k == i: frequency_dictionary[k]+=1/len(countries_dataframe)
    
    list_replace_attributes = []
    for i in frequency_dictionary.keys():
        if frequency_dictionary[i]<percentage/100:
            list_replace_attributes.append(i)
    
    return countries_dataframe.replace(to_replace=list_replace_attributes,value='GENERIC')
```

```python
naive_dimensionality_reduction(countries_dataframe,percentage=24)
```

## TASK 2 b)
Why could this function be useful in building a logistic regression model predicting a binary target like the Delayed (y/n) in the above dataset?

By replacing the least expressive (in terms of frequency) samples and aggregating them into a new attribute, a naive type of dimensionality reduction is performed. Since we do not have to consider N additional variables to represent the least expressive samples, the most expressive variables are captured more accurately, making a logistic regression model to generalize better, instead of trying to overfit to all samples.

## TASK 2 c)
Describe shortly another approach, which could be useful to treat a categorical attribute with too many different values.

Instead of replacing the least expressive samples and then perform one-hot encoding on the dataset, one could perform a custom binary encoding and choose a specific one that would reason better, if there was enough knowladge about the problem at hand. For instance, if we were performing regression on the price of a car, and had many categorical variables to describe it, such as brand, one could aggregate expensive brands and least expressive brands by making a custom encoding (expensive cars -- one variable)/(medium cars -- one variable)/(cheap cars- one variable).

Alternatively, describe briefly an algorithm that addresses internally
this issue and does not require such a treatment of the categorical attributes.

Instead of applying the methods mentioned above, one could apply robust dimensionality reduction methods to handle the problem automatically. One of the simpler methods, corresponds to PCA which can be solely derived by linear algebra and is equivalent to applying an orthogonal transformation to the dataset. The eigenvectors and eigenvalues of the covariance matrix determine the directions and magnitude of the transformation. This transformation corresponds to capturing the axis (principal components) matching the highest variance of the data. As was stated by Jonathon Shlens: "Performing PCA is like choosing a camera angle, to gain the best possible view of the variance to be explained.".

However, PCA is not able to deal with non-linearly correlated data and one needs to "drop the camera and move the data itself". In this sense there are many state-of-the-art methods such as k-means and more recently with Deep Learning, deep auto-encoders, which achieved significantly state-of-the-art results in handwritten digits (MNIST dataset).
